```python
# Here we just make release_ data as the index column
df.set_index("release_date",inplace=True)
df.index=pd.to_datetime(df.index,errors='coerce',infer_datetime_format=True)
df.head()
```

```python
# Check if 'duration_ms' column exists before proceeding
if 'duration_ms' in df.columns:
    df["duration"]=df["duration_ms"].apply(lambda x : round(x/1000))
    df.drop("duration_ms",inplace=True,axis=1)
else:
    print("Column 'duration_ms' not found in DataFrame.")
```

```python
corra_df = df.select_dtypes(include=['number']).corr(method="pearson")
plt.figure(figsize=(14,4))
heatmap=sns.heatmap(corra_df,annot=True,fmt=".1g",vmin=-1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df.sample(int(0.004*len(df)))
# thid gives the 4% of my total data frame
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(16,4))
plote=sns.regplot(data=sample_df,y="loudness",x="energy",color="c")
plote.set(title="loudness vs Energy correlatiob")
# this is the relaation in loudness and energy
```

```python
plt.figure(figsize=(16,4))
plote=sns.regplot(data=sample_df,y="popularity",x="acousticness",color="r")
plote.set(title="popularity vs acousticness correlatiob")
# this is the relaation in popularity and acousticness
```

```python
# lets find number of songs released per year
# Check if 'release_date' column exists before processing
if 'release_date' in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'])
    years = df['release_date'].dt.year
    sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per years")
else:
    print("Column 'release_date' not found in the DataFrame.")
```

```python
# lets find number of songs released per year
df = df.reset_index()# as i just reset the dateset to it contain the release_date
df['release_date'] = pd.to_datetime(df['release_date'])
years = df['release_date'].dt.year
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per years")
```
